#### imports numpy, astropy.table, tabulate, and defines x

```python
import numpy as np
from tabulate import tabulate
from astropy.table import Table
x = np.linspace(0, 2 * np.pi, 1000)
```

#### This is the function that calculates *sin(x)*:

```python
def sin_fxn(x):
    return np.sin(x)
```

#### This is the function that calculates *cos(x)*:

```python
def cos_fxn(x):
    return np.cos(x)
```

#### This is the cell that tabulates *sin(x)* and *cos(x)*:

```python
table = Table()
table['x'] = x
table['sin(x)'] = sin_fxn(x)
table['cos(x)'] = cos_fxn(x)
print(table)
```

#### The for loop that prints the first 10 values of *x*, *sin(x)*, and *cos(x)*:

```python
for i in range(10):
    print(f"{x[i]:15.14f}, {sin_fxn(x)[i]:15.14f}, {cos_fxn(x)[i]:15.14f}")
```
